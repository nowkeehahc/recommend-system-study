```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read movies.csv file and print the data shape and first few rows
movies_df = pd.read_csv('movies.csv')
print('Movies data shape:', movies_df.shape)
print(movies_df.head())
```

```python
# Count movies by genre and plot the result
genres_count = movies_df['genres'].str.split('|').apply(pd.Series).stack().value_counts()
genres_count.plot(kind='bar')
plt.title('Number of Movies by Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.show()
```

```python
# Read ratings.csv file and print the data shape and first few rows
ratings_df = pd.read_csv('ratings.csv')
print('Ratings data shape:', ratings_df.shape)
print(ratings_df.head())
```

```python
user_ids = ratings_df['userId'].unique()
print('Number of Users:', len(user_ids))
```

```python
# Plot rating distribution
ratings_df['rating'].hist()
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Ratings')
plt.show()
```

```python
# Read tags.csv file and print the data shape and first few rows
tags_df = pd.read_csv('tags.csv')
print('Tags data shape:', tags_df.shape)
print(tags_df.head())
```

```python
# Count top 10 tags and plot the result
tag_count = tags_df['tag'].value_counts().head(10)
tag_count.plot(kind='bar')
plt.title('Top 10 Tags')
plt.xlabel('Tag')
plt.ylabel('Number of Tags')
plt.show()
```
