```python
import pandas as pd
import scipy.sparse as sp
```

```python
# Read the ratings.csv file
ratings_df = pd.read_csv('ratings.csv')
```

```python
# Create the user-item matrix
user_item_matrix = pd.pivot_table(ratings_df, values='rating', index='userId', columns='movieId')
user_item_matrix
```

```python
# Convert the user-item matrix to a sparse matrix
user_item_matrix_sparse = sp.csr_matrix(user_item_matrix.fillna(0).values)
user_item_matrix_sparse
```
